```python
import thinkdsp
import thinkplot

%matplotlib inline
```

1,2) Now that we've loaded up the module, we can load in a sample sound. This is a recording from the buzzer on my guitar tuner playing an A note.

```python
wave = thinkdsp.read_wave('ps01.wav').segment(start=1, duration=0.1)
wave.segment(duration=0.01).plot()
wave.make_audio()
```

As can be shown above, the signal is pretty periodic. In fact, it looks like a bunch of jagged sinusoids.

3) Next, we'll plot the spectrum to see what components make up this signal.

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=1000)
for f, amp in zip(spectrum.fs, spectrum.amps):
    if amp > 500000:
        print(f, amp)
```

So, we can see that the dominant frequency in this signal is at 1320 Hz which is 3 times the A note frequency of 440 Hz. Note that 1320 Hz is not an even octave of A (the octaves are at 220, 440, 880, 1760 Hz).

4) Next, let's try clobbering all of the frequencies above 1000 Hz to see if there is a fundamental buried somewhere in there.

```python
spectrum = wave.make_spectrum()
spectrum.low_pass(1000,0.01)
for f, amp in zip(spectrum.fs, spectrum.amps):
    if amp > 50000:
        print(f, amp)
spectrum.make_wave().make_audio()
```

That was kind of cool. Note the frequencies at 220, 440 and 880 Hz, all lower octaves of the same A note. It sounds like a lower version of the same note, which it is.

Now, let's try the opposite: cutting out everything except the higher frequencies.

```python
spectrum = wave.make_spectrum()
spectrum.high_pass(2500,0.01)
spectrum.low_pass(5000,0.01)
spectrum.plot(low=250, high=500)
spectrum.make_wave().make_audio()
```

Now, this sounds like an A note of a higher octave. That's kind of cool.

5) Now, let's create an arbituary wave. We start by adding three random harmonics. 

```python
wave = thinkdsp.Sinusoid(amp=0).make_wave()
spectrum = wave.make_spectrum()
for _ in xrange(3):
    spectrum.hs[thinkdsp.numpy.random.randint(0,5400)] = thinkdsp.numpy.random.ranf()
spectrum.plot()
spectrum.make_wave().make_audio()
```

That's pretty random sounding. Let's add a bunch more random ones:

```python
for _ in xrange(9):
    spectrum.hs[thinkdsp.numpy.random.randint(0,5400)] = thinkdsp.numpy.random.ranf()
spectrum.plot()
spectrum.make_wave().make_audio()
```

Now, it just sounds more muddled. If we add enough, we end up at static:

```python
for _ in xrange(200):
    spectrum.hs[thinkdsp.numpy.random.randint(0,5400)] = thinkdsp.numpy.random.ranf()
spectrum.plot()
spectrum.make_wave().make_audio()
```

Eww, static.

6) Now, let's make a class to generate a sawtooth wave. This is rather similar to the sinuosid class except for in how it calculates the ys so we will inherit that class.

```python
class Sawtooth(thinkdsp.Sinusoid):
    def evaluate(self, ts):
        ys = -(ts % self.period)
        return thinkdsp.normalize(ys, self.amp)
  
    
signal = Sawtooth()
wave = signal.make_wave()
wave.segment(duration=0.01).plot()
```

7) Let's check out the spectrum of this wave:

```python
spectrum = wave.make_spectrum()
spectrum.plot()
for f, amp in zip(spectrum.fs, spectrum.amps):
    if amp > 100:
        print(f, amp)
```

Voila! This wave has components at the natural frequency and then at all odd harmonics. We can compare this to the triangle wave:

```python
signal = thinkdsp.TriangleSignal()
wave = signal.make_wave()
spectrum = wave.make_spectrum().plot()
```

And the square wave:

```python
signal = thinkdsp.SquareSignal()
wave = signal.make_wave()
spectrum = wave.make_spectrum().plot()
```

Compared to the triangle wave, the sawtooth wave has more harmonics and the higher harmonics don't drop off quite as quickly. The square wave drops off at a similar rate to the sawtooth wave but also has less harmonics.

8) Next, we'll look for aliasing in the triangle wave.

```python
signal = thinkdsp.TriangleSignal(freq=1100)
wave = signal.make_wave(framerate=10000)
wave.make_spectrum().plot()
wave.make_audio()
```

Yikes. You can hear that aliased harmonic and it is nasty. In this case, the aliased harmonic we are seeing is the one around 4500 which should actually be at 5500 Hz.

9) Let's try a square wave:

```python
signal = thinkdsp.SquareSignal(freq=1100)
wave = signal.make_wave(framerate=10000)
wave.make_spectrum().plot()
wave.make_audio()
```

Still sounds pretty nasty. In this case, the peak around 3300 should not be there.

10) We can make a wave with even and odd harmonics that drop off like $1/f^2$:

```python
wave = thinkdsp.Sinusoid(amp=0).make_wave()
spectrum = wave.make_spectrum()

for i in xrange(10):
    freq = 440*(i+1)
    spectrum.hs[freq] = 1./(freq**2)
    
spectrum.make_wave().segment(duration=0.01).plot()
spectrum.make_wave().make_audio()
```

That's a pretty exciting wave... Haven't seen that before.
